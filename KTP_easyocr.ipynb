```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
import difflib
import pandas as pd
import os
from data_extract import extract_data, mapping_data
import re


class KtpProcessor:
    def __init__(self):
#         self.list1 = ["rt/rw", "nama", "alamat", "NIK", "tempat tanggal lahir", "gol. darah", "jenis kelamin", "kel/desa", "kecamatan", "agama", "status perkawinan", "pekerjaan"]
        self.reader = easyocr.Reader(['en'], gpu=True)
    
    def read_ktp(self, file_paths,save_path='ktp.xlsx'):
        data_list = []
        
        for file in os.listdir(file_paths):
            if file.endswith('.jpg') or file.endswith('.png'):
                img_path = os.path.join(file_paths, file)
                result = self.reader.readtext(img_path)
                raw_data = [x[1].split(':') for x in result]
                correct_data = [self.autocorrect(j.lower()) for i in raw_data for j in i]
                
                extracted = extract_data(correct_data)
                mapped_data = mapping_data(extracted)
                data_list.append(mapped_data)
                
            #
            
        ktp_file = pd.DataFrame(data_list)
        ktp_file.to_excel(excel_writer=save_path, sheet_name='Data_Ktp')
        
        return ktp_file
    
    def input_ktp(self, indeks, correct_data):
        data_ktp = []

        for i in indeks:
            data_ktp.append(correct_data[i+1])

        return data_ktp
    
    def autocorrect(self, input_data):
        # Find the closest matching word from the vocabulary
        vocabulary = ["rt/rw", "nama", "alamat", "NIK", "tempat tanggal lahir", "gol. darah", "jenis kelamin", "kel/desa", "kecamatan", "agama", "status perkawinan", "pekerjaan"]
        
        closest_match = difflib.get_close_matches(input_data, vocabulary, n=1, cutoff=0.6)

        if closest_match:
            return closest_match[0]
        else:
            return input_data
            
```

```python
IMAGE_PATH = './Samples/ktp/'

tool = KtpProcessor()
tool.read_ktp(file_paths=IMAGE_PATH)
```
